## Univaritate , Bivaritate and Multivariate analysics

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv")
df.head()
```

-- We have 3 species in the dataset setosa,virgina and vesicolor

```python
df.shape
```

## Univariate analysics -- Take only one feature (here we are considering the petal length) and see with the help of that feature we are able to classify the points or not

```python
# Making a seprate dataframe for all the 3 species


df_setosa=df.loc[df['species']=='setosa']
df_versicolor=df.loc[df['species']=='versicolor']
df_virginica=df.loc[df['species']=='virginica']
```

```python
df_virginica.head(n=5)
```

```python
# We are plotting univariate graph so we will keep all our y values as 0



# np.zeros_like takes length as the argument
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'x')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'^')
plt.xlabel('Petal_Length')
plt.show()
```

```python
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'x')
plt.xlabel('Petal_Length')
plt.show()
```

## Bivariate_Analysics

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,"petal_length","sepal_length").add_legend()
```

## We can see that we can classify the above points visually

## Multivariate (to visulaize more than 3 features)

```python
sns.pairplot(df,hue='species',size=2)    # This is basically the size of the picture
```

## From the above we can infer that petal length and petal width are the most important feature for model buliding
